```python
from PIL import Image 
import numpy as np
```

```python
IMAGE_PATH = "Nancy_pilosi.jpg"
OUTPUT_WIDTH=100
```

```python
# ASCII characters from dark to light
ASCII_CHARS= ["*", "#", "@", " "]
```

```python
img= Image.open(IMAGE_PATH)
```

```python
img=img.convert("L")  # Convert to grayscale
```

```python
#resize image maintaining aspect ratio
width, height=img.size
aspect_ratio= height/width
char_aspect= 0.55 
new_height= int(OUTPUT_WIDTH * aspect_ratio * char_aspect)
img= img.resize((OUTPUT_WIDTH, new_height))
```

```python
img= img.convert("L")  
```

```python
# resize image maintaining aspect ratio

width, height=img.size
aspect_ratio= height/width
char_aspect= 0.55
new_height= int(OUTPUT_WIDTH * aspect_ratio * char_aspect)

img= img.resize((OUTPUT_WIDTH, new_height))
```

```python
# Image to numpy array
img_array= np.array(img)
```

```python
# Map pixel to ASCII
bins=np.linspace(0, 256, len(ASCII_CHARS)+1)
ascii_indices= np.digitize(img_array, bins) - 1
ascii_indices= np.clip(ascii_indices, 0, len(ASCII_CHARS)-1)
```

```python
ascii_art=[]
for row in ascii_indices:
    ascii_art.append("".join([ASCII_CHARS[i] for i in row]))
```

```python
print("\n".join(ascii_art))
```
